```python
#Finding the values which have anomalies in HSSC-1 column


#Finding out all non-numeric enteries and enteries greater than maximum marks   

# for mark in df['HSSC-1']:
#     if mark.isdigit():
#         if int(mark) <= 510:
#             pass
#         else:
#             print(mark)
#     else:
#         print(mark)


# Removing all the HSSC-1 marks that are above 510, because they can not be scaled
df2 = df
marks = []
results = []

#df2.drop(df2[(pd.to_numeric(df2['HSSC-1']) > 510)].index, inplace=True)
# df2

for name, gender, mark, part2, gpa, fav_color, month, weight in zip(df2['Name'], df2['Gender'], df2['HSSC-1'], df2['HSSC-2'], df2['CGPA'], df2['FavoriteColor'], df2['BirthMonth'], df2['Weight']):
    if mark.isdigit():
        if int(mark) <= 510:
            marks.append(mark)
            jsonObj = {"Name":name, "Gender": gender, "HSSC-1": mark, "HSSC-2": part2, "CGPA": gpa, "FavoriteColor": fav_color, "BirthMonth": month, "Weight": weight}
            results.append(jsonObj)
        else:
            print(mark)
    else:
        if mark == "60%":
            marks.append(int(60/100*510))
            jsonObj = {"Name":name, "Gender": gender, "HSSC-1": int(60/100*510), "HSSC-2": part2, "CGPA": gpa, "FavoriteColor": fav_color, "BirthMonth": month, "Weight": weight}
            # df2['HSSC-1'] = df2['HSSC-1'].replace('60%', int(60/100*510))
            results.append(jsonObj)
        elif mark == "368/510":
            marks.append(368)
            jsonObj = {"Name":name, "Gender": gender, "HSSC-1": 368, "HSSC-2": part2, "CGPA": gpa, "FavoriteColor": fav_color, "BirthMonth": month, "Weight": weight}
            results.append(jsonObj)

# len(marks)
# pd.set_option('display.max_rows', None)

#Creating a new data frame without HSCC-1 marks greater 510
df3 = pd.DataFrame(results)
df3
```

```python
#Finding the values which have anomalies in HSSC-2 column

#Finding out all non-numeric enteries and enteries greater than maximum marks

# for mark in df3['HSSC-2']:
#     if mark.isdigit():
#         if int(mark) <= 590:
#             pass
#         else:
#             print(mark)
#     else:
#         print(mark)

#If entery is greater than maximum marks, subtracting the total marks with part 1 marks
#Correcting the invalid type of enteries but contain correct marks
marks = []
for mark1, mark2 in zip(df3['HSSC-1'], df3['HSSC-2']):
    if mark2.isdigit():
        if int(mark2) <= 590:
            marks.append(mark2)
        else:
            if mark1.isdigit():
                marks.append(int(mark2) - int(mark1))
            else:
                print(mark2)
    else:
        if mark2 == "768/1100":
            marks.append(768 - int(mark1))
        else:
            print(mark2)
        
part2 = marks
```

```python
#Finding all the non-numeric enteries of weight
# for weight in df3['Weight']:
#     if weight.isdigit():
#         pass
#     else:
#         print(weight)
        
#Formatting the incorrect values
weights = []
for weight in df3['Weight']:
    if weight.isdigit():
        weights.append(weight)
    else:
        x = re.findall(r'\d+', weight)
        a = x[0]
        weights.append(a)

# len(weights)
```

```python
def checkMonth(month):
    if month == "January" or month == "February" or month == "March" or month == "April" or month == "May" or month == "June" or month == "July" or month == "August" or month == "September" or month == "October" or month == "November" or month == "December":
        return True
    else:
        return False
    
incorrect_values = 0
months = []
for birth_month in df3['BirthMonth']:
    if checkMonth(birth_month.strip().capitalize()) == True:
        months.append(birth_month.strip().capitalize())
    else:
        if birth_month.strip().capitalize() == 'Feburary':
            months.append("February")
        elif birth_month.strip() == 'sep':
            months.append("Spetember")
        elif birth_month == "11":
            months.append("November")
        elif birth_month == "24/02/2001":
            months.append("Feburary")
        else:
            print(birth_month)
            incorrect_values +=1
        

# print(incorrect_values)    
# print(len(months))

# correct_values = 0
# for i in months:
#     if checkMonth(i) == True:
#         correct_values+=1
        
```

```python
gpas = []

def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False


for gpa in df3['CGPA']:
    if isfloat(gpa) == True or gpa.isdigit() == True:
        gpas.append(gpa)
    else:
        x = gpa.removesuffix('.')
        gpas.append(x)
        print(gpa)
        
# print(len(gpas))
```

```python
colors = []

for color in df3['FavoriteColor']:
    # y = re.findall(r'\w+\s\w+\s\w+$', color) 
    x = color.strip().lower().capitalize()
    colors.append(x)

# myset = set(colors)
# print(myset)
# li = list(myset)
# print(li)

print(len(colors))
```

```python
name = df3['Name']
part1 = df3['HSSC-1']
gender = df3['Gender']
```

```python
dictionary = {'Name':name,
            'Gender': gender, 
            'HSSC-1': part1,
            'HSSC-2': part2,
            'CGPA': gpas,
            'FavoriteColor': colors,
            'BirthMonth': months,
            'Weight': weights,  
        }

df4 = pd.DataFrame(dictionary)
df4
```
