```python
from bs4 import BeautifulSoup as Soup
import requests
import pandas as pd
import json
```

```python
url = "https://www.shophive.com/sale" 

resp = requests.get(url) 

resp = resp.text 
```

```python
soup = Soup(resp, 'html.parser')
```

```python
lis = soup.findAll('li', {'class':'product-item'})
```

```python
results = []
for li in lis:
    name = li.find('a', {'class':'product-item-link'}).text
    price = li.find('span', {'class':'price'}).text
    img = li.find('img', {'class':'lazyOwl lazy product-image-photo'})['data-original']
    jsonObj = {"name":name, "price":price, "link":img}
    results.append(jsonObj)
```

```python
results
```

```python
df = pd.DataFrame(results)
df
```

```python
df.to_csv("test.csv")
```
